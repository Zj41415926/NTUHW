```python
# 當日買權履約價與成交價
import pandas as pd
url='http://www.taifex.com.tw/cht/3/optDailyMarketSummary'
c= pd.read_html(url)[4]
c = c.drop(c.columns[5:9],axis=1)
c=c.drop(c.columns[2:4],axis=1)
c.columns =("買權到期月份(週別)","履約價","成交價")
c.drop([0],inplace=True)

c
```

```python
## Final Code 輸入日期來爬取期交所選擇權成交價
import string
import requests
import re
from bs4 import BeautifulSoup 

def findoptionvalue(date):
    payload ={
    'queryType': '2',
    'marketCode': '0',
    'dateaddcnt': '',
    'commodity_id': 'TXO',
    'commodity_id2': '',
    'queryDate': format(date),
    'MarketCode': '0',
    'commodity_idt': 'TXO',
    'commodity_id2t':'' ,
    'commodity_id2t2':'',
    }
    rs = requests.session()
    res=rs.post('http://www.taifex.com.tw/cht/3/optDailyMarketReport',data = payload)
    soup = BeautifulSoup(res.text,'lxml') 
    m = re.findall('(<td class="12bk"><div align="center">.*</td>)',res.text)
    
    if  len(m):
        for tr in  soup.select('.section tr')[11:455]:
            td = tr.select('td')
            print(td[0].text,td[1].text,td[2].text,td[3].text,td[7].text)   
    else:
        print("今日休市")
        
        
        
findoptionvalue('2018/12/28') #輸入在這
```
